```python
def add(a, b):
    return a + b

def substract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Division by zero"
    return a / b
    
```

```python
print("Calculator")
print("1. Add")
print("2. Substract")
print("3. Multiply")
print("4. Divide")
```

```python
choice = input("choose 1/2/3/4 only")
x = float(input("Enter first number: "))
y = float(input("Enter second number: "))

if choice == "1":
    print("Result: ", add(x, y))
if choice == "2":
    print("Result: ", substract(x, y))
if choice == "3":
    print("Result: ", multiply(x, y))
if choice == "4":
    print("Result: ", divide(x, y))
```
